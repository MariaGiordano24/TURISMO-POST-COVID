```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carico il file Excel.
df_europa = pd.read_excel('DATI_EUROPEI_PERNOTTAMENTI.xlsx')
df_europa
```

```python
#Elimino le colonne che iniziano con 'Unnamed'
df_europa = df_europa.loc[:, ~df_europa.columns.str.contains('^Unnamed')]
df_europa.head()
```

```python
#Elimino la riga 0 che contiene "GEO (Labels)" 
df_europa = df_europa[df_europa['TIME'] != 'GEO (Labels)']

#Faccio il MELT (Unpivot)
df_europa_long = df_europa.melt(id_vars=['TIME'], var_name='Data', value_name='Pernottamenti')
df_europa_long.head()
```

```python
#Rinomino la colonna da TIME a Paese
df_europa_long = df_europa_long.rename(columns={'TIME': 'Paese'})
df_europa_long.head()
```

```python
#Elimino eventuali spazi bianchi all'inizio o alla fine dei nomi.
df_europa_long['Paese'] = df_europa_long['Paese'].astype(str).str.strip()
```

```python
#Creo il dizionario.
traduzione_paesi = {
    'Germany': 'Germania',
    'Spain': 'Spagna',
    'France': 'Francia',
    'Italy': 'Italia'
}

df_europa_long['Paese'] = df_europa_long['Paese'].map(traduzione_paesi)
df_europa_long.head()
```

```python
#Trasformo pernottamenti in numeri interi.
df_europa_long['Pernottamenti'] = df_europa_long['Pernottamenti'].astype(int)
df_europa_long.head()
```

```python
df_europa_long.info()
```

```python
#Trasformo la colonna Data da testo a "Vera Data". 
df_europa_long['Data'] = pd.to_datetime(df_europa_long['Data'])
df_europa_long.info()
```
