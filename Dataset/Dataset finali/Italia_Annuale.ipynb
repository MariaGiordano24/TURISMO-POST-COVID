```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Mi carico il file Excel.
df = pd.read_excel('PERNOTTAMENTI_ITALIA_REGIONI_2019_2024.xlsx')
df
```

```python
#Uso .shape per vedere quante colonne e righe ci sono.
df.shape
```

```python
#Uso .info per vedere le informazioni generali del dataset con relativo tipo di dato.
df.info()
```

```python
#Mi tengo solo le colonne che mi servono e non vuote.
df = df.loc[:, ~df.columns.str.contains('^Unnamed')] 
#Controllo che non ci siano le colonne vuote.
df.head()
```

```python
#Elimino la riga 0 che contiene "GEO (Labels)"
df = df[df['TIME'] != 'GEO (Labels)']

#Faccio il MELT ovvero (l'Unpivot) trasformando la tabella da orizzontale a verticale.
df_long = df.melt(id_vars=['TIME'], var_name='Anno', value_name='Pernottamenti')

#Controllo il risultato finale.
df_long.head(15)
```

```python
#Controllo anche le ultime righe.
df_long.tail(5)
```

```python
#Ricontrollo quante righe e colonne.
df_long.shape
```

```python
#Rinomino la colonna TIME in Regione.
df_long = df_long.rename(columns={'TIME': 'Regione'})

#Visualizzo le prime righe per vedere se il nome è cambiato.
df_long.head()
```

```python
#Tolgo gli spazi vuoti all'inizio e alla fine dei nomi delle regioni.
df_long['Regione'] = df_long['Regione'].str.strip()

#Controllo i nomi unici delle regioni.
df_long['Regione'].unique()
```

```python
#Sostituisco i nomi delle regioni.
df_long['Regione'] = df_long['Regione'].replace({
    'Valle d’Aosta/Vallée d’Aoste': "Valle d'Aosta",
    'Provincia Autonoma di Bolzano/Bozen': 'Trentino-Alto Adige',
    'Provincia Autonoma di Trento': 'Trentino-Alto Adige'
})

#Controllo che siano stati cambiati. 
df_long['Regione'].unique()
```

```python
#Sommo i valori di Bolzano e Trento.
df_long = df_long.groupby(['Regione', 'Anno'], as_index=False)['Pernottamenti'].sum()

#Ordino per Regione e per Anno.
df_long = df_long.sort_values(by=['Regione', 'Anno'])
```

```python
#Ricontrollo che le righe siano diminuite.
df_long.shape
```

```python
#Trasformo i Pernottamenti in numeri interi (senza il .0).
df_long['Pernottamenti'] = df_long['Pernottamenti'].astype(int)

#Controllo il risultato.
df_long.head()
```

```python
#Controllo che la colonna Pernottamenti abbia il tipo di dato diverso ovvero da float a int.
df_long.info()
```

```python
#Salvo
df_long.to_excel('PERNOTTAMENTI_REGIONI_ITALIA_2019_2024_PULITO.xlsx', index=False)
```
